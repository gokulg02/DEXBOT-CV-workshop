```python
def prespective_transform(img):
    source = np.float32([[7,149],[255,10],[321,329],[533,91]])
    destination = np.float32([[0,0],[300,0],[0,300],[300,300]])
    transformMatrix = cv2.getPerspectiveTransform(source, destination)
    transformedImg = cv2.warpPerspective(img,transformMatrix,(300,300))
    return transformedImg
```

```python
f= cv2.imread(r'data/vid_img/frame43.jpg')
pt_img = prespective_transform(f)
plt.imshow(pt_img[:,:,::-1])
```

```python
def conv_hsv(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    return img
```
