```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
```

__Manual Thresholding__

```python
img = cv2.imread(r"data\boat.jpg")
```

```python
img.shape
```

```python
img = img[:,:,0]
```

```python
plt.imshow(img, cmap = 'gray')
```

```python
plt.hist(np.ravel(img))
```

```python
mask = img < 150
```

```python
plt.imshow(mask, cmap = 'gray');
```

```python
img_thresh = mask*img
```

```python
f, ax = plt.subplots((2),figsize=(10, 10))
ax[0].imshow(img, cmap = 'gray');
ax[1].imshow(img_thresh, cmap = 'gray');
```

__Automatic Thresholding__

```python
ret,mask = cv2.threshold(img,0,1,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
```

```python
plt.imshow(mask, cmap = 'gray');
```

```python
img_thresh = mask*img
```

```python
f, ax = plt.subplots((2),figsize=(10, 10))
ax[0].imshow(img, cmap = 'gray');
ax[1].imshow(img_thresh, cmap = 'gray');
```
