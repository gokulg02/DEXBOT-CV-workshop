__Image Convolutions__

```python
img = cv2.imread(r'data/robin.jpeg')
kernel1 = np.array([[0, 0, 0],
                    [0, 1, 0],
                    [0, 0, 0]])
identity = cv2.filter2D(img, -1, kernel1)
f, ax = plt.subplots(2,figsize=(10, 10))
ax[0].imshow(img[:,:,::-1])
ax[1].imshow(identity[:,:,::-1])
```

__Blurring__

```python
kernel1 = np.ones((5,5))/25
blur = cv2.filter2D(img, -1, kernel1)
f, ax = plt.subplots(2,figsize=(10, 10))
ax[0].imshow(img[:,:,::-1])
ax[1].imshow(blur[:,:,::-1])
```

```python
#Using inbuilt function
img_blur = cv2.blur(img, (10,10))
f, ax = plt.subplots(2,figsize=(10, 10))
ax[0].imshow(img[:,:,::-1])
ax[1].imshow(img_blur[:,:,::-1])
```

__Gaussian Blurring__

```python
gaussian_blur = cv2.GaussianBlur(img, (11,11),sigmaX=0, sigmaY=0)
f, ax = plt.subplots(2,figsize=(10, 10))
ax[0].imshow(img[:,:,::-1])
ax[1].imshow(gaussian_blur[:,:,::-1])
```

```python
f, ax = plt.subplots(2,figsize=(10, 10))
ax[0].imshow(img_blur[:,:,::-1])
ax[1].imshow(gaussian_blur[:,:,::-1])
ax[0].set_title("Average blur")
ax[1].set_title("Gaussian blur")
```

__Sharpening__

```python
kernel1 = np.array([[0, -1, 0],
                    [-1, 5, -1],
                    [0, -1, 0]])
img_sharp = cv2.filter2D(gaussian_blur, -1, kernel1)
f, ax = plt.subplots(2,figsize=(10, 10))
ax[0].imshow(gaussian_blur[:,:,::-1])
ax[1].imshow(img_sharp[:,:,::-1])
```
